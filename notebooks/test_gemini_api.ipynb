# Gemini API Key Test\nQuick sanity check that `GEMINI_API_KEY` is set and working.

```python
from pydantic_settings import BaseSettings

class Settings(BaseSettings):
    gemini_api_key: str

    class Config:
        env_file = "../.env"
        extra = "ignore"

settings = Settings()
print(f"GEMINI_API_KEY found ({len(settings.gemini_api_key)} chars, starts with {settings.gemini_api_key[:8]}...)")
```

```python
from google import genai
from google.genai import types

client = genai.Client(api_key=settings.gemini_api_key)

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=["Say 'Hello! Gemini is working.' and nothing else."],
    config=types.GenerateContentConfig(
        temperature=0.0,
        max_output_tokens=20,
        thinking_config=types.ThinkingConfig(thinking_budget=0),
    ),
)

# Debug: inspect the full response if .text is None
if response.text:
    print(response.text)
else:
    print("response.text is None — inspecting candidates:")
    for candidate in response.candidates:
        for part in candidate.content.parts:
            print(f"  part: {part}")
```
